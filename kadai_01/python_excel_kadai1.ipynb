```python
import openpyxl
import datetime

# ワークブックとシートを新しく作成
workbook = openpyxl.Workbook()
ws = workbook.active

# 現在時刻を取得する
now = datetime.datetime.now()

ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'
ws['F4'] = 'No.'
ws['F5'] = '日付'
ws['G4'] = '0001'

# 日付を代入
ws['G5'] = now.strftime('%Y/%m/%d')

# ヘッダーを'B10'セルから書き込む
header = ['商品名', '数量', '単価', '金額']
header_row = 10
for j, h_text in enumerate(header):
    # B列は2列目なので column=2+j
    ws.cell(row=header_row, column=2 + j, value=h_text)

# 複数行のデータをリストdataに格納
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000],
    [ '', '', '小計', 35000],
]

# '商品A'を含むデータを'B11'セルから入力を開始
start_row = 11
for i, row_data in enumerate(data):
    for j, value in enumerate(row_data):
        # B列(2列目)から順に入力
        ws.cell(row=start_row + i, column=2 + j, value=value)

# 合計欄のセル位置もデータに合わせて調整
ws['B15'] = '合計'
ws['B16'] = '消費税'
ws['B17'] = '税込合計'
ws['E15'] = 35000
ws['E16'] = 3500
ws['E17'] = 38500

# ファイル名の保存
filename = f"請求書_{now.strftime('%Y%m%d')}.xlsx"
workbook.save(filename)
print(f"{filename} を作成しました。")
```
